```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType

constructors_schema = StructType([
  StructField("constructorId", IntegerType(), False),
  StructField("constructorRef", StringType(), False),
  StructField("name", StringType(),  False),
  StructField("nationality", StringType(), True),
  StructField("url", StringType(), False),
])
```

```python
constructors_df = spark.read.json("dbfs:/mnt/f1dlnagsa/raw/constructors.json", schema=constructors_schema)
```

```python
from pyspark.sql.functions import current_timestamp
constructors_final_df = constructors_df.drop("url")\
        .withColumnRenamed("constructorRef", "constructor_ref")\
        .withColumnRenamed("constructorId", "constructor_id")\
        .withColumn("ingestion_date",current_timestamp())
```

```python
constructors_final_df.write.mode("overwrite").format("delta").saveAsTable("f1_processed.constructors")
```
